# Análisis de Ventas — Tienda

This notebook analyzes sales data from two store datasets. It covers:
- Data loading and inspection
- Type conversion and cleaning
- Merging datasets
- Key business insights: best-selling products and top revenue months

## 1. Imports

```python
import pandas as pd
from pathlib import Path
```

## 2. Load Data

Two CSV files are loaded — update the paths below to match your local setup.

```python
path1 = Path("Datos_Ventas_Tienda.csv")
path2 = Path("Datos_Ventas_Tienda2.csv")

df_tienda1 = pd.read_csv(path1)
df_tienda2 = pd.read_csv(path2)

df_tienda1.head(3)
```

## 3. Data Inspection

```python
print(f"Tienda 1 shape: {df_tienda1.shape}")
print(f"Tienda 2 shape: {df_tienda2.shape}\n")

df_tienda1.info()

print("\nNull values:")
print(df_tienda1.isnull().sum())
```

## 4. Data Cleaning

The `Fecha` (date) column is read as `object` — convert it to `datetime` in both datasets.

```python
df_tienda1["Fecha"] = pd.to_datetime(df_tienda1["Fecha"])
df_tienda2["Fecha"] = pd.to_datetime(df_tienda2["Fecha"])

print(df_tienda1["Fecha"].dtype)  # Expected: datetime64[ns]
```

## 5. Merge Datasets

```python
df_total = pd.concat([df_tienda1, df_tienda2], ignore_index=True)

print(f"Combined dataset shape: {df_total.shape}")
df_total.head()
```

## 6. Analysis

### 6.1 Best-Selling Product (by units)

```python
producto_mas_vendido = (
    df_total.groupby("Producto")["Cantidad"]
    .sum()
    .sort_values(ascending=False)
)

print("Top 5 products by units sold:")
producto_mas_vendido.head()
```

### 6.2 Best Month by Revenue

```python
df_total["Mes"] = df_total["Fecha"].dt.month

ventas_por_mes = (
    df_total.groupby("Mes")["Total Venta"]
    .sum()
    .sort_values(ascending=False)
)

print("Revenue by month (top 5):")
ventas_por_mes.head()
```
